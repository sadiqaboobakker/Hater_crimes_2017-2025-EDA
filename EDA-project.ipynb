```python


#   8  "How has the percentage distribution of hate crime offenders by age group (Under 18, 18+, Unknown) changed between 2017-2023 and 2024-2025?"-





print("\nOffender Age Trends in 2024-2025 vs 2017-2023:")
df['Number of Offenders under 18'] = pd.to_numeric(df['Number of Offenders under 18'], errors='coerce').fillna(0)
df['Number of Offenders over 18'] = pd.to_numeric(df['Number of Offenders over 18'], errors='coerce').fillna(0)
recent = df[df['Year'].isin([2024, 2025])]
previous = df[df['Year'].isin(range(2017, 2024))]
def age_group(row):
    if row['Number of Offenders under 18'] > 0:
        return 'Under 18'
    elif row['Number of Offenders over 18'] > 0:
        return '18+'
    else:
        return 'Unknown'
recent['Offender Age Group'] = recent.apply(age_group, axis=1)
previous['Offender Age Group'] = previous.apply(age_group, axis=1)
previous_age = previous['Offender Age Group'].value_counts(normalize=True) * 100
recent_age = recent['Offender Age Group'].value_counts(normalize=True) * 100
plt.show()
print("\nAge Group Distribution (%):")
```

```python
age_df = pd.DataFrame({
    '2017-2023': previous_age,
    '2024-2025': recent_age
}).fillna(0) 
plt.figure(figsize=(10, 6))
age_df.plot(kind='bar', color=['#1f77b4', '#ff7f0e'], width=0.8)
plt.title('Offender Age Group Distribution Comparison', pad=20)
plt.ylabel('Percentage of Incidents (%)')
plt.xlabel('Age Group')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.3)
plt.legend(title='Time Period')
plt.tight_layout()
for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height():.1f}%', 
                      (p.get_x() + p.get_width() / 2., p.get_height()), 
                      ha='center', va='center', 
                      xytext=(0, 5), 
                      textcoords='offset points')
```

```python



#   conclusion


#   Adult offenders (18+) consistently represent the majority of perpetrators in both time periods, though the exact percentage shows [increase/decrease/stability] from [X]% (2017-2023) to [Y]% (2024-2025).

```

```python


#  9 How did the COVID-19 pandemic period (2020-2021) affect the monthly patterns of hate crimes compared to other years?


import pandas as pd
import matplotlib.pyplot as plt

df['Date of Incident'] = pd.to_datetime(df['Date of Incident'])
df['Year'] = df['Date of Incident'].dt.year
df['Month'] = df['Date of Incident'].dt.month_name()

pre_covid_df = df[df['Year'] < 2020]
covid_df = df[(df['Year'] >= 2020) & (df['Year'] <= 2021)]

pre_covid_monthly = pre_covid_df.groupby('Month').size().reset_index(name='Count_pre')
covid_monthly = covid_df.groupby('Month').size().reset_index(name='Count_covid')

monthly_comparison = pre_covid_monthly.merge(covid_monthly, on='Month', how='outer').fillna(0)

monthly_comparison['Pct_Change'] = (
    (monthly_comparison['Count_covid'] - monthly_comparison['Count_pre']) /
    monthly_comparison['Count_pre'].replace(0, float('nan'))
) * 100

monthly_comparison['Month'] = pd.Categorical(
    monthly_comparison['Month'],
    categories=['January', 'February', 'March', 'April', 'May', 'June',
                'July', 'August', 'September', 'October', 'November', 'December'],
    ordered=True
)
monthly_comparison = monthly_comparison.sort_values('Month')

pre_mean = pre_covid_monthly['Count_pre'].mean()
covid_mean = covid_monthly['Count_covid'].mean()
change_pct = ((covid_mean - pre_mean) / pre_mean) * 100

print("\nMonthly Comparison Statistics:")
print(f"Pre-COVID average monthly incidents: {pre_mean:.1f}")
print(f"COVID period average monthly incidents: {covid_mean:.1f}")
print(f"Change during COVID: {change_pct:+.1f}%")

print("\nMonth-by-Month Percentage Changes (%):")
print(monthly_comparison[['Month', 'Pct_Change']].set_index('Month').T.round(1))

```

```python
plt.figure(figsize=(12, 6))
bar_width = 0.4
index = range(len(monthly_comparison))

plt.bar(index, monthly_comparison['Count_pre'], bar_width, label='Pre-COVID', color='gray')
plt.bar([i + bar_width for i in index], monthly_comparison['Count_covid'], bar_width, label='COVID (2020-21)', color='skyblue')
plt.xticks([i + bar_width / 2 for i in index], monthly_comparison['Month'], rotation=45)
plt.xlabel('Month')
plt.ylabel('Number of Incidents')
plt.title('Monthly Hate Crime Incidents: Pre-COVID vs COVID Period')
plt.legend()
plt.tight_layout()
plt.show()
plt.figure(figsize=(10, 4))
plt.plot(monthly_comparison['Month'], monthly_comparison['Pct_Change'], marker='o', color='red')
plt.axhline(0, color='black', linestyle='--')
plt.title('Percentage Change in Monthly Hate Crime Incidents (COVID vs Pre-COVID)')
plt.xlabel('Month')
plt.ylabel('Percentage Change (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python



#     conclusion




#    The data shows a [increase/decrease/no significant change] in average monthly hate crimes during the pandemic period (2020-2021) compared to pre-pandemic years (2017-2019), with a [X]% change overall


```
